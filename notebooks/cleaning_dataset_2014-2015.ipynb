```python
#Import librairies 
import pandas as pd
import numpy as np
import re
import cleaning_functions
```

```python
#Load the dataset
df = pd.read_csv('../data/raw/2014_2015.csv', encoding='ISO-8859-1')
df.head()
```

```python
#Cleaning the dataset using the cleaning_functions.py created beforehand
#Using in order :

# 1/ Droping columns that have no interest for the project 
# 2/ Clean column names - converting to snake format + manual transformation
# 3/ Converting some columns values to snake format 
# 4.1/ Converting rating note to a float number 
# 4.2/ Converting rating note to a 3 level scale (positive/neutral/negative)
# 5/ Manual cleaning of the values in the 'price' column, in order to get only the numerical values
# 6/ Converts all values in the 'category' column to lowercase
# 7/ Splits the 'category' column at each '/' and adds new columns to structure it in main_category and sub_categories
# 8.1/ Manual cleaning for the values in the 'origin' column 
# 8.2/ Manual cleaning for the values in the 'destination' column
# 9/ Manual cleaning of some satelite values
# 10/ Save cleaned dataframe as a clean_dataset.csv
```

```python
#1

columns_to_drop = [' Remarks']
df = cleaning_functions.drop_columns(df, columns_to_drop)
df.head()
```

```python
#2

df = cleaning_functions.clean_column_names(df)
df.head()
```

```python
df.columns
```

```python
df = df.rename(columns={
    '_category' : 'category',
    '_item': 'item',
    '_item_description': 'item_description',
    '_price': 'price',
    '_origin': 'origin',
    '_destination': 'destination',
    '_rating': 'rating'
})
df.columns
```

```python
#3

df = cleaning_functions.value_snake_case(df, ['vendor','origin', 'destination']) 
df.head()
```

```python
#4.1

from cleaning_functions import extract_number

df['numeric_rating'] = df['rating'].apply(extract_number)
df.head()
```

```python
#4.2

from cleaning_functions import categorize_rating

df['rating_category'] = df['numeric_rating'].apply(categorize_rating)
df.head()
```

```python
#5
df['price'] = df['price'].str.extract(r'(\d+\.\d+)').astype(float)
df
```

```python
#6

df = cleaning_functions.lowercase_categories(df)
df
```

```python
#7

df = cleaning_functions.split_category(df)
df.rename(columns={0: 'main_category', 1: 'sub_category'}, inplace=True)
df.head()
```

```python
df['category'].unique()
```

```python
df['main_category'].unique()
```

```python
#drop irelevant categories
unwanted_categories = ['the body bags  shotty  and mag to: kind hearts and gentle people', 
                      '0.12780125125 btc',  '0.1905617980645162 btc', 
                       'home of the body bags  shotty  and mag  kind hearts and gentle people' ]
df = df[~df['category'].isin(unwanted_categories)]
df = df[~df['main_category'].isin(unwanted_categories)]
```

```python
#boolean masks for each category group
is_drug_paraphernalia = df['main_category'].str.contains('drug paraphernalia', case=False)
is_forgeries = df['main_category'].str.contains('^forgeries', case=False)
is_info_data = df['main_category'].str.contains('^(info|data)', case=False)
is_other = df['main_category'].str.contains('^other', case=False)

# update the values in the category column
df.loc[is_drug_paraphernalia, 'main_category'] = 'drugs'
df.loc[is_forgeries, 'main_category'] = 'counterfeits'
df.loc[is_info_data, 'main_category'] = 'data'
df.loc[is_other, 'main_category'] = 'services'
```

```python
#8.1 'origin'
#too many messy values
unique_origins = df['origin'].value_counts().reset_index()
unique_origins.columns = ['origin', 'count']
unique_origins.head(50)
```

```python
#8.1 'origin'
#let's get the counts of each unique value
counts = df['origin'].value_counts()
# Let's identify the values that occur less than 125 times
less_than_125 = counts[counts < 125].index
#let's replace these values with NaN
df['origin'] = df['origin'].map(lambda x: np.nan if x in less_than_125 else x)

unique_origins = df['origin'].value_counts().reset_index()
unique_origins.columns = ['origin', 'count']
unique_origins.head(50)
```

```python
#8.1 'origin'
##let's replace ['eu', 'torland', 'world', 'internet', 'undeclared','nan'] values by NaN
df['origin'] = df['origin'].replace(['eu', 'torland', 'internet', 'undeclared'], np.nan)

unique_origins = df['origin'].value_counts().reset_index()
unique_origins.columns = ['origin', 'count']
unique_origins.head(50)
```

```python
df.head()
```

```python
#8.2 'destination'
#too many messy values
unique_destination = df['destination'].value_counts().reset_index()
unique_destination.columns = ['destination', 'count']
unique_destination.head(50)
unique_destination.head(50)
```

```python
#8.2 'destination'
#let's get the counts of each unique value
counts = df['destination'].value_counts()
# Let's identify the values that occur less than 140 times
less_than_125 = counts[counts < 140].index
#let's replace these values with NaN
df['destination'] = df['destination'].map(lambda x: np.nan if x in less_than_125 else x)

unique_destination = df['destination'].value_counts().reset_index()
unique_destination.columns = ['destination', 'count']
unique_destination.head(50)
```

```python
#8.2
##let's replace several values
replacements = {
    'usa_only': 'usa',
    'world_wide': 'world',
    'worldwide': 'world',
    'worldwide_except_usa': 'world',
    'everywhere': 'world',
    'european_union': np.nan,
    'worldwide_except_australia': 'world',
    'canada': 'usa',
    'international': 'world',
    'eu_only': np.nan,
    'eu': np.nan,
    'australia_only': 'australia',
    'united_states': 'usa',
    'usa_canada': 'usa',
    'uk_only': np.nan,
    'europe_union_only': np.nan
}

df['destination'] = df['destination'].replace(replacements)


unique_destination = df['destination'].value_counts().reset_index()
unique_destination.columns = ['destination', 'count']
unique_destination.head(50)
```
